# Explore data: lets search for variables that I may need for my analysis

```python
#Examples to just search for what we want from the catalog
expname_filter = ['SPEAR_c192_o1_Scen_SSP585_IC2011_K50_ens_01_03']
realm = 'atmos'
frequency = "mon"
variable_id = ["ps","psl","hfss","hfls","rlus","rlds","rsus","rsds","snow","evap","lwup_sfc","lwdn_sfc","swup_sfc","swdn_sfc","t_surf","WVP"]
```

```python
cat_mon = col.search(experiment_id=expname_filter,
                 realm=realm,frequency=frequency)
```

```python
cat_mon.df.realm.unique()
```

```python
#cat.df[['variable_id','standard_name']]
```

```python
cat_mon = col.search(experiment_id=expname_filter,
                 realm=['ocean','ocean_z'],frequency=frequency)
```

```python
#cat.df[['variable_id','standard_name']]
#cat_mon.df[['path']]
```

```python
day6_variable_id = ["psl","pr","prs","tas","ts","uas","vas","tasmax","tasmin","huss","hurs","swe",
                     "hfss","hfls","rlus","rlds","rsus","rsds","slp","ps","precip","snow_tot","t_ref","wind_ref","u_ref","v_ref","t_ref_max",
                     "t_ref_min","q_ref","rh_ref","snow","evap","lwup_sfc","lwdn_sfc","swup_sfc","swdn_sfc","t_surf","WVP"]
cat_daily = col.search(experiment_id=expname_filter,realm='atmos_daily',
                frequency='day',variable_id=day6_variable_id)
```

```python
cat_daily.df[['path']]
```

```python
cat_6hr = col.search(experiment_id=expname_filter,
                frequency='6hr',variable_id=day6_variable_id)
```

```python
cat_6hr.df["path"]
```

```python
catm = (cat_mon.df["path"].to_frame())
catd = (cat_daily.df["path"].to_frame())
cat6 = (cat_6hr.df["path"].to_frame())
```

```python
import csv
catm.to_csv('/home/Aparna.Radhakrishnan/git/sandbox/cats/spear/select.txt',header=None, index=None, sep=' ', mode='a')
catd.to_csv('/home/Aparna.Radhakrishnan/git/sandbox/cats/spear/select.txt',header=None, index=None, sep=' ', mode='a')
cat6.to_csv('/home/Aparna.Radhakrishnan/git/sandbox/cats/spear/select.txt',header=None, index=None, sep=' ', mode='a')
```
